```python
import pandas as pd
import json
from tqdm import trange
```

# video metadata to csv

```python
with open("taitra_all/taitra_yt.txt",'r') as file:
    df = json.loads(file.read())
file.close()
video = pd.DataFrame(df[1:],columns=df[0])
video
```

```python
df = pd.read_csv("taitra_all/表格資料_edit_thumbnails.csv")

indus_list = [None, '化學品', '電腦', '五金', '消費性電子', '汽機車', '農產及食品', '機械', '運動用品',
       '機械零件', '電機及電子', '美容及個人用品', '禮品及藝品', '工具', '電信', '自行車', '服飾及配件',
       '船舶及漁業', '量測及分析儀器', '建材及衛浴設備', '家庭用電器', '樂器', '家庭用品', '家具', '航太',
       '醫療及保健', '鐘錶、珠寶及眼鏡', '礦產及金屬材料', '包裝、印刷及廣告', '', '辦公室及學校用品',
       '燈光及照明']
video_list = [None, 'edm', 'poll', 'tutorial']

indus_types = []
indus_num = []
video_types = []
video_num = []
for i in trange(len(df)):
    if df.iloc[i]["id"] in video["直式影片 ID"].values:
        mask = video["直式影片 ID"]==df.iloc[i]["id"]
        indus_type = video[mask]["產業類別"].values[0]
        video_type = video[mask]["選擇版型"].values[0].split("_")[1]
        
        indus_types.append(indus_type)
        indus_num.append(indus_list.index(indus_type))
        video_types.append(video_type)
        video_num.append(video_list.index(video_type))
        
    elif df.iloc[i]["id"] in video["橫式影片 ID"].values:
        mask = video["橫式影片 ID"]==df.iloc[i]["id"]
        indus_type = video[mask]["產業類別"].values[0]
        video_type = video[mask]["選擇版型"].values[0].split("_")[1]
        
        indus_types.append(indus_type)
        indus_num.append(indus_list.index(indus_type))
        video_types.append(video_type)
        video_num.append(video_list.index(video_type))
    else:
        indus_types.append(None)
        indus_num.append(0)
        video_types.append(None)
        video_num.append(0)

df['product'] = indus_types
df["product_encode"] = indus_num
df["form"] = video_types
df["form_encode"] = video_num
df.to_csv("taitra_all/表格資料_edit_thumbnails.csv", index=False)
```
